```python
import collections

def bfs(grid, start, end):
    height, width = grid.shape[:2]
    queue = collections.deque([[start]])
    seen = set([start])
    while queue:
        path = queue.popleft()
        x, y = path[-1]
        if x == end[0] and y == end[1]:
            return np.array(path)
        for x2, y2 in [(x+1,y), (x-1,y), (x,y+1), (x,y-1), (x+1,y+1), (x-1,y-1), (x-1,y+1), (x+1,y-1)]:
            if 0 <= x2 < width and 0 <= y2 < height and grid[y2,x2] > 0 and (x2, y2) not in seen:
                queue.append(path + [(x2, y2)])
                seen.add((x2, y2))
    return None
```

```python
animal = 'zebra'
input_dir = 'data/preprocessed/%s/' % animal

masks = np.load(input_dir + 'masks.npy')
feat_img = np.load(input_dir + 'feat_img.npy')
feat_part = np.load(input_dir + 'feat_part.npy')

print(masks.shape)
print(feat_img.shape)
    
bs = masks.shape[0]
hw = masks.shape[-1]

for i in range(bs):
    labels, n_feat = ndimage.label(masks[i,0])
    max_component = 1
    n_pixels = np.sum(labels == 1)
    for c in range(2, n_feat+1):
        if np.sum(labels == c) > n_pixels:
            n_pixels = np.sum(labels == c)
            max_component = c
    masks[i,0] *= labels == max_component
```

```python
plt.figure(figsize=(10,12))

img_xy = np.arange(hw).astype(np.float32)
grid_y, grid_x = np.meshgrid(img_xy, img_xy, indexing='ij')

feat_cons = np.ones((8))
feat_all = np.zeros((bs,8))
for i in range(bs):
    sil_x = grid_x[masks[i,0]>0]
    cent_x = np.mean(sil_x)
    for j in range(8):
        m = masks[i,0] == j+1
        if np.sum(m) == 0:
            feat_cons[j] = 0
        else:
            part_cent_x = np.mean(grid_x[m])
            # cluster_var = np.mean(np.sqrt((grid_x[m] - part_cent_x)**2))
            feat_all[i,j] = (part_cent_x - cent_x) # / cluster_var


feat_mean = np.mean(np.abs(feat_all), axis=0)
# face_cluster = np.argmax(feat_mean*feat_cons)
face_cluster = np.argmax(feat_mean)
print(face_cluster+1)


rot = np.ones(bs).astype(np.float32)
for i in range(bs):
    if feat_all[i,face_cluster] < 0:
        feat_all[i] *= -1
        masks[i,0,:,:] = masks[i,0,:,::-1]
        rot[i] *= -1

indices = np.argsort(np.mean(np.abs(feat_all), axis=1))
feat_all = feat_all[indices]
masks = masks[indices]
for i in range(bs):
    plt.subplot(6,5,i+1)
    plt.imshow(masks[i,0], cmap='tab10')
    # plt.imshow(masks[i,0]==face_cluster, cmap='tab10')
    plt.axis('off')
    plt.colorbar()
    
cam_rot = np.arange(bs).astype(np.float32)/(bs-1)
cam_rot *= np.pi/2
print([rot[i]*cam_rot[np.where(indices==i)[0][0]] for i in range(bs)])
    
plt.tight_layout()
```

```python
remove_joints = [9]

for j in remove_joints:
    joints_3d = torch.cat([joints_3d[:j], joints_3d[j+1:]], 0)
    for i, p in enumerate(joints_parent):
        if p == j:
            joints_parent[i] = joints_parent[j]
        if p > j:
            joints_parent[i] -= 1
    joints_parent = joints_parent[:j] + joints_parent[j+1:]
    part_dino = np.array(list(part_dino[:j]) + list(part_dino[j+1:]))
    part_scale_xz = torch.cat([part_scale_xz[:j], part_scale_xz[j+1:]], 0)
    part_scale_y = torch.cat([part_scale_y[:j], part_scale_y[j+1:]], 0)

print(joints_3d.shape[0]-1)
print(joints_3d)
print(joints_parent)
print(list(part_dino[1:]-1))
print(part_scale_xz.cpu()[1:])
print(part_scale_y.cpu()[1:])


plt.figure(figsize=(10,10))
for c, p in enumerate(joints_parent):
    plt.plot([joints_3d[c,2].cpu(), joints_3d[p,2].cpu()], [joints_3d[c,1].cpu(), joints_3d[p,1].cpu()])
plt.scatter(joints_3d[:,2].cpu(), joints_3d[:,1].cpu(), c=np.arange(joints_3d.shape[0]), vmin=0, vmax=20, cmap='tab20')
plt.colorbar()
```

```python
merge_pairs = [(4,5)]

for j1, j2 in merge_pairs:
    joints_3d[j1] = (joints_3d[j1] + joints_3d[j2])/2
    for i, p in enumerate(joints_parent):
        if p == j2:
            joints_parent[i] = j1

remove_joints = [j2]
for j in remove_joints:
    joints_3d = torch.cat([joints_3d[:j], joints_3d[j+1:]], 0)
    for i, p in enumerate(joints_parent):
        if p == j:
            joints_parent[i] = joints_parent[j]
        if p > j:
            joints_parent[i] -= 1
    joints_parent = joints_parent[:j] + joints_parent[j+1:]
    part_dino = np.array(list(part_dino[:j]) + list(part_dino[j+1:]))
    part_scale_xz = torch.cat([part_scale_xz[:j], part_scale_xz[j+1:]], 0)
    part_scale_y = torch.cat([part_scale_y[:j], part_scale_y[j+1:]], 0)
            
        
print(joints_3d.shape[0]-1)
print(joints_3d)
print(joints_parent)
print(list(part_dino[1:]-1))
print(part_scale_xz.cpu()[1:])
print(part_scale_y.cpu()[1:])


plt.figure(figsize=(10,10))
for c, p in enumerate(joints_parent):
    plt.plot([joints_3d[c,2].cpu(), joints_3d[p,2].cpu()], [joints_3d[c,1].cpu(), joints_3d[p,1].cpu()])
plt.scatter(joints_3d[:,2].cpu(), joints_3d[:,1].cpu(), c=np.arange(joints_3d.shape[0]), vmin=0, vmax=20, cmap='tab20')
plt.colorbar()
```

```python
symmetry_pairs = [(10,11)]

for j1, j2 in symmetry_pairs:
    joints_3d[j1,0] = -0.05
    joints_3d[j2,0] = 0.05
    joint_mean = (joints_3d[j1] + joints_3d[j2])/2
    for c, p in enumerate(joints_parent):
        if p == j1:
            joints_3d[c,1:] += joint_mean[1:] - joints_3d[j1,1:]
            joints_3d[c,0] = joints_3d[j1,0]
        elif p == j2:
            joints_3d[c,1:] += joint_mean[1:] - joints_3d[j2,1:]
            joints_3d[c,0] = joints_3d[j2,0]
    joints_3d[j1,1:], joints_3d[j2,1:] = joint_mean[1:], joint_mean[1:]
    part_scale_xz_mean = (part_scale_xz[j1] + part_scale_xz[j2])/2
    part_scale_y_mean = (part_scale_y[j1] + part_scale_y[j2])/2
    part_scale_xz[j1], part_scale_xz[j2] = part_scale_xz_mean, part_scale_xz_mean
    part_scale_y[j1], part_scale_y[j2] = part_scale_y_mean, part_scale_y_mean
    
    parent = joints_parent[j1]
    if sum([p == parent for p in joints_parent]) == 2:
        joints_parent = [p+1 if p > parent else p for p in joints_parent]
        joints_parent[j2] += 1
        joints_parent.insert(parent, joints_parent[parent])

        joints_3d = torch.cat([joints_3d[:parent+1], joints_3d[parent:]], 0)
        joints_3d[parent,0] = -0.05
        joints_3d[parent+1,0] = 0.05

        part_scale_xz[parent] *= 0.5
        part_scale_xz = torch.cat([part_scale_xz[:parent+1], part_scale_xz[parent:]], 0)
        part_scale_y = torch.cat([part_scale_y[:parent+1], part_scale_y[parent:]], 0)
        part_dino = np.concatenate((part_dino[:parent+1], part_dino[parent:]))

print(joints_3d.shape[0]-1)
print(joints_3d)
print(joints_parent)
print(list(part_dino[1:]-1))
print(part_scale_xz.cpu()[1:])
print(part_scale_y.cpu()[1:])


plt.figure(figsize=(10,10))
for c, p in enumerate(joints_parent):
    plt.plot([joints_3d[c,2].cpu(), joints_3d[p,2].cpu()], [joints_3d[c,1].cpu(), joints_3d[p,1].cpu()])
plt.scatter(joints_3d[:,2].cpu(), joints_3d[:,1].cpu(), c=np.arange(joints_3d.shape[0]), vmin=0, vmax=20, cmap='tab20')
plt.colorbar()
```
